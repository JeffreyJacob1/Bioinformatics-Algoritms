# Problem 8: 

```python
#Rosalind problem 8
#inputs a string of nucleotides
# returns the Kmer Composition of a given string
# Groupmates: Alan Faber, Sarah Ghasemi
# Author: Jeffrey Jacob

#Oct 17, 2022

class problemEight():
    """ Description: This class prints the kmers in the sequence """
 
    
    def __init__ (self, kmerLen, sequence):
        """ Description: This method initializes variables to be used within the class, the inputs
        are the length of the kmers and the sequence itself"""

        self.kmerLen = kmerLen #Initializes the length of the kmer as self
        self.sequence = sequence 
        self.totalLength = len(sequence) 

    def printKmers (self):
        """ Description: The print kmers method prints all the kmers in the sequence 
        the output is a alphabetically sorted list of possible kmers"""
        Kmers = []
        for i in range(int(self.totalLength)-(int(self.kmerLen))+1): #Goes over all lengths of the string that can produce kmers
            Kmers.append(self.sequence[i:i+(int(self.kmerLen))]) #Kmer is appended to a list Kmers
        for i in (sorted(Kmers)): #Sorts Kmers alphabetically
            print(i)

def main(File):
    """ Description: Utilizes methods in the class to accept a file as a input and output the sorted list of Kmers"""
    with open(File, "r") as file:
        length = file.readline().rstrip() 
       
        sequence = file.readline().rstrip() 
    val = problemEight(length, sequence) 
    val.printKmers() #Indentifies and prints out the kmers in the sequence

if __name__== "__main__":
    main("rosalind_ba3a.txt")
```

```python
#Rosalind problem 11
#inputs: a string
# returns: the de Brujin adjacency list
# Groupmates: Alan Faber, Sarah Ghasemi
# Author: Jeffrey Jacob

#Oct 17, 2022
class problemEleven():
    """ This class models a object, in this case a string text and a integer k"""
    def DeBruijn(self,text, k):
        """ This is the main function which creates the adjacency list corresponding to the DeBruijn list"""
        
        result = {}
        
        for i in range(len(text)-k+1):
            if text[i:i+k-1] not in result.keys():
                result[text[i:i+k-1]] = text[i+1:i+k] #Creating keys and values that correspond the the adjacency list
            else:
                result[text[i:i+k-1]] += ',' + text[i+1:i+k] #case where one key has multiple values
        return result

def main(inFile):
    
    """ Main part of program where file in inputted and class methods are called which print the adjacency list"""
    
    with open(inFile, 'r') as f:
        k = int(f.readline().strip())
        
        text = f.readline().strip()
    graph = problemEleven()   
    r = graph.DeBruijn(text, k) #calling method
    for key in sorted(r.keys()):
        print(key + ' -> ' + r[key])
        
if __name__== "__main__":
    main("rosalind_ba3d.txt")
```

```python
#Rosalind problem 12
#inputs: a string
# returns: the de Brujin adjacency lis
# Groupmates: Alan Faber, Sarah Ghasemi
# Author: Jeffrey Jacob

class problemTwelve():
    """ Description: This class constructs the De Bruijn Graph of a Collection of k-mers """
    def __init__(self, seqList):
        """ Description: The method initializes variables as self  It accepts input of a list of sequences"""
        self.seqList = seqList 

    def nodeDict(self):
        """ Description: This method creates a dictionary of nodes, it outputs said dictionary"""
        nodes = {}
        for i in self.seqList: #Iterates through sequences and saves as []
            nodes[i[:-1]] = []
        for i in self.seqList: #Iterates through sequences and saves as correpsonding nodes
            nodes[i[:-1]].append(i[1:])
        return nodes #Returns the resulting dictionary

def main(inFile):
    """ Description: The main prints out the De Bruijn Graph of a Collection of k-mers"""
    """ Input: The input is a text file containing sequences """
 
    file = open(inFile, "r")
    seqSet = []
    for line in file:
        seqSet.append(line.rstrip()) #Saves each sequence into a list
    processed = problemTwelve(seqSet) #Initializes the variables
    nodeDict = processed.nodeDict() #Creates a dictionary from sequences
    for i in sorted(nodeDict): #Sorts the dictionary and prints it out in a certain fashion
        print(i + ' -> ' + ','.join(sorted(nodeDict[i])))

if __name__== "__main__":
    main("rosalind_ba3e.txt")

```

```python
#Rosalind problem 11
#inputs: an integer k followed by kmer patterns
# returns: a string with kmer-comp equal to the pattern
# Groupmates: Alan Faber, Sarah Ghasemi
# Author: Jeffrey Jacob


import random



class Graph:
    """  The class Graph takes a set of DNA sequences of size k and constructs the nodes and edges needed to make the sequence comp"""  
  
    
    def __init__(self, patterns):
        """initalization method for the class self.patterns is the patterns inputted """
        self.patterns = patterns
        self.edges = [] # list of tuples representing edges for the graph, generated by self.findEdges()
        self.nodes = dict() # dictionary in which each key is a pattern and each value is a tuple containing two lists: the
        self.stringPath = [] # list of patterns in the Eulerian path, generated by self.findPath()  first containing the incoming nodes, the second containing the outgoing nodes
        self.sequence = '' # the outputted sequence composition of all kmers within patterns

    def findEdges(self):
        """
        Finds all the edges between the kmers within self.patterns by comparing the (k-1) length suffixes and prefixes
        of each kmer. no inputs or outputs """
        for seqPattern in self.patterns: #If the suffix of one pattern is equal to the prefix of another pattern, then a tuple containing  both patterns is added to the self.edges.
            for otherPattern in self.patterns:
                if seqPattern[1:] == otherPattern[:-1]:
                    self.edges.append((seqPattern, otherPattern))

   
    def findNodes(self):
        """  Uses self.edges generated in self.findEdges() to generate the dictionary self.nodes. 
        """
        self.findEdges()  
        for edge in self.edges: #For every edge in self.pattern, each pattern contained in edge is added to the self.node
            node1 = edge[0]
            node2 = edge[1]# Add an outgoing edge to node1 (edge[0])
            if node1 not in self.nodes.keys(): #If a pattern shows up  in the first position of the tuple (node1), then we update the pattern in the second position (node2)
                self.nodes[node1] = ([], [])
           
            self.nodes[node1][1].append(node2)  
            if node2 not in self.nodes.keys():
                self.nodes[node2] = ([], [])
            self.nodes[node2][0].append(node1)

    def getStart(self):
        """ Finds the start node of the Eulerian path. The start node of an Eulerian path is a node whose number of outgoing
        edges number of incoming edges is equal to 1 This is accomplished by going through all the nodes in self.nodes and subtracting the length of the outgoing nodes 
        to the in coming nodes"""
        self.findNodes()
        startNode = random.choices(list(self.nodes.keys()))[0]
        for node in self.nodes.keys():
            if len(self.nodes[node][1]) - len(self.nodes[node][0]) == 1: #when out - in == 1
                startNode = node
        return startNode

    def findPath(self, v):
        """ accepts the param v, which is the graphs starting node and uses recursion to find Eulerian Cycles in the graph
        while there are still outgoing nodes we randomly remove one and call findPath again with the selected node. This is added to the front of the list"""
        vEdges = self.nodes[v][1] #out edges
        while len(vEdges) > 0:
            newV = random.choices(vEdges)[0]
            self.nodes[v][1].remove(newV)
            self.findPath(newV)
        self.stringPath.insert(0, v)

    def printString(self):
        """
        Using the patterns within self.stringPath, prints the generated sequence.
        Calls on self.findPath() using the node outputted from self.getStart().
        """
        self.findPath(self.getStart())
        for i in range(0, len(self.stringPath)):
            if i == 0:
                self.sequence += self.stringPath[i]
            else:
                self.sequence += self.stringPath[i][-1]
        print(self.sequence)


def main(inFile):
    patterns = []
    lines = 0
    fileRead = open(inFile, 'r')
    # Read the input file
    for line in fileRead:
        if lines > 0:
            patterns.append(line.strip())
        lines += 1

    seqGraph = Graph(patterns)
    seqGraph.printString()


if __name__ == "__main__":
    main('rosalind_ba3h.txt')
```
